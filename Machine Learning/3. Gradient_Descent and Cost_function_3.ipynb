```python
import pandas as pd

df=pd.read_csv('test_scores.csv')
df
```

```python
import numpy as np
import math


x=np.array(df[['math']])
y=np.array(df[['cs']])
```

```python
def gradient_descent(x,y):     # We have to adjust learning_rate first with small iteration values
    m_curr = b_curr = 0        # After that, we have to maximize iteration
    iterations = 1000000
    n = len(x)
    learning_rate = 0.0002
    prev_cost=0
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr               # y=mx+b for current values
        cost_f = (1/n) * sum([val**2 for val in (y-y_predicted)]) # cost function/ mean square error  
        pd_m = -(2/n)*sum(x*(y-y_predicted))            # Partial derivative of m
        pd_b = -(2/n)*sum(y-y_predicted)                #  Partial derivative of b
        m_curr = m_curr - learning_rate * pd_m          # calculating current value of m with learning rate
        b_curr = b_curr - learning_rate * pd_b          # calculating current value of b with learning rate
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost_f, i)) # printing current values of each iteration
        if math.isclose(cost_f,prev_cost,rel_tol=1e-20):
            break
        prev_cost=cost_f
gradient_descent(x,y)
```

### So, Value of m=1.01773817, b=1.91508262 and iteration it took 4,15,533
