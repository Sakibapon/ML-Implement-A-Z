### Solving using Pandas Dummy Variables

```python
dummies = pd.get_dummies(df['Car Model'])    # Dummies of town column
dummies
```

```python
merged_df=pd.concat([df,dummies],axis='columns')
merged_df
```

```python
final_df=merged_df.drop(['Car Model'],axis='columns')
final_df
```

```python
final_df=final_df.drop(['Mercedez Benz C class'],axis='columns') # Avodinig Dummy trapX=
final_df
```

```python
X=final_df.drop(['Sell Price($)'],axis='columns')  # Sell Price dependent variable
X
```

```python
y=final_df['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,y)
```

```python
model.predict(X) # Predicting prices for all X
```

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,0,0]])  # price of a mercedez benz that is 4 yr old with mileage 45000
```

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])  # price of a BMW X5 that is 7 yr old with mileage 86000
```

#### The score (accuracy) of model

```python
model.score(X,y)   # 94.17% accuracy
```
