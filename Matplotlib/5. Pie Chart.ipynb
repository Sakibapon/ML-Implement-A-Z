# Pie Chart
#### Usually used to represent percentage of data

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)  # pie chart to track down home expenses
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True) # adding shadow
plt.axis("equal")           # axis equal makes perfect circle
plt.show()
```

```python
plt.axis("equal")                     # Radius for increasing/decreasing circle size
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.2f%%',radius=1.5) 
plt.show()                             # aautpct for automatically percentage represenataion
```

### Explode

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()       # Explode takes arrays similar to values of pie chart array
                 # By default it's 0. Increasing from 0 will make come out pie of that labels
```

### Rotating and changing position of pie

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()                # Counterclock and startangle
```
