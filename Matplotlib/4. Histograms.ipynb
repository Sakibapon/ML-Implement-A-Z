# Histograms

### *Difference between Bar and Histogram is-* 
#### 1. Histograms are used to show distributions of variables while bar charts are used to compare variables.
#### 2. A histogram represents the frequency distribution of continuous variables. Conversely, a bar graph is a diagrammatic comparison of discrete variables. 
#### 3. Histogram presents numerical data whereas bar graph shows categorical data.
#### 4. x axis contains a variable and y axis will be a frequency of that variable

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129,140] # blood sugar sample of 14 people
plt.hist(blood_sugar) # by default number of bins is set to 10 (how many bar in histogram)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129,140]
plt.hist(blood_sugar, rwidth=0.8) # rwidth= relative width of bar compare to bin size
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)     # bins parameter
```

### Histogram showing normal, pre-diabetic and diabetic patients distribution
#### *0-100: Normal*
#### *100-125: Pre-diabetic*
#### *more than 125: Diabetic*

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

### Mutiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [112, 140, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 100, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

#### histtype

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step') #hist type step
```

#### orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal') # Horizontal orientation
```
