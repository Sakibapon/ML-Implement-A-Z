# Format String Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]    #initializing values
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--') # Format String as 3rd argument
```

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)   # markersize
```

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1. Its indicate transparency
```
