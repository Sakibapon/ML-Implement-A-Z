# Bar Chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company)) # matplotlib does not allow string as axis. So, converting to numpy array
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)      # xticks rplacing x array values with labels
plt.ylabel("Revenue(Billion)")
plt.title('US Technology Stocks')
plt.legend()
```

### Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]  # new array for bar chart
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue") # xpos of two bar chart same. so, substracting for better view
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")    # Total width of single bar=1. It's 0.4 for avoid overlapping

plt.xticks(xpos,company)
plt.ylabel("Revenue(Billion)")
plt.title('US Technology Stocks')
plt.legend()
```

### Horizontal bar chart

```python
plt.barh(xpos,revenue, label="Revenue")   #barh

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
